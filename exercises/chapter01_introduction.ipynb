## 問題8: 関数の作成

以下の関数を作成してください。

1. 摂氏温度を華氏温度に変換する関数 `celsius_to_fahrenheit`
   - 変換式: F = C × 9/5 + 32
2. 作成した関数を使って、25度（摂氏）を華氏に変換

````r
# 問題8の解答欄

````

### 模範解答

````r
# 模範解答
celsius_to_fahrenheit <- function(celsius) {
  fahrenheit <- celsius * 9/5 + 32
  return(fahrenheit)
}

# 25度（摂氏）を華氏に変換
celsius_to_fahrenheit(25)
````
